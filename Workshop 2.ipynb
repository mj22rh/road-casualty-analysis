```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('./MCU.db')
curr = conn.cursor()
```

```python
table_cmd = \
"""
create table MCU_Actors(
    actor_id int PRIMARY_KEY,
    actor_name text,
    movie_id int, 
    FOREIGN KEY (movie_id) REFERENCES MCU_Movies(movie_id) 
);

"""
```

```python
curr.execute(table_cmd)
conn.commit()
```

```python
actor_data = [
(1, 'Robert Downey Jr.', 1001),
(2, 'Chris Evans', 1005),
(3, 'Mark Ruffalo', 1006),
(4, 'Chris Hemsworth', 1004),
(5, 'Scarlett Johansson', 1006),
(6, 'Jeremy Renner', 1006),
(7, 'Tom Hiddleston', 1004),
(8, 'Clark Gregg', 1001),
(9, 'Paul Rudd', 1012),
(10, 'Benedict Cumberbatch', 1014)
]
# Insert actor data into the MCU_Actors table
curr.executemany("INSERT OR IGNORE INTO MCU_Actors VALUES(?,?,?)", actor_data)
conn.commit()
```

```python

actors = curr.execute("select * from MCU_Actors")
actors = actors.fetchall()
```

```python
actors_df = pd.DataFrame(actors)
titles = {0:'actor_id', 1:'actor_name', 2:'movie_id'}
actors_df.rename(columns = titles, inplace = True)
actors_df
```

```python
cmd_join = \
"""
select MCU_Movies.*,MCU_Actors.actor_name 
from MCU_Movies 
join MCU_Actors  on MCU_Actors.movie_id = MCU_Movies.movie_id
"""

movies = curr.execute(cmd_join)
movies = movies.fetchall()
```

```python
movie_df = pd.DataFrame(movies)
movie_titles = {0:'movie_id', 1:'title', 2:'year', 3: 'score', 4:'actor_name'}
movie_df.rename(columns= movie_titles, inplace=True)
movie_df
```
